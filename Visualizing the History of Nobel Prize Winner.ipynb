The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)



```python
# libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
Dataset=pd.read_csv("nobel.csv")
Dataset.head()
```

```python
top_country=Dataset.value_counts('birth_country').index[0]
top_country
```

```python
top_gender=Dataset.value_counts("sex").index[0]
top_gender
```

```python
Dataset['usa_born_winner']=Dataset['birth_country'] == 'United States of America'
Dataset.head()
```

```python
Dataset['decade']= (np.floor(Dataset['year']/10)*10).astype(int)
Dataset
```

```python
Proportion_US=Dataset.groupby('decade',as_index=False)['usa_born_winner'].mean()
Proportion_US
```

```python
max_decade_usa=Proportion_US[Proportion_US['usa_born_winner']==Proportion_US['usa_born_winner'].max()]
max_decade_usa
```

```python

ax1=sns.relplot(x='decade',y='usa_born_winner',data=Proportion_US,kind='line')
ax1.fig.suptitle("Decade vs Proportion USA")

```

```python
Dataset['Female_winner']= Dataset['sex']=='Female'
Dataset
```

```python
Proportion_Female_Winner=Dataset.groupby(['decade','category'],as_index=False)['Female_winner'].mean()
Proportion_Female_Winner
```

```python
max_female_decade_category=Proportion_Female_Winner[Proportion_Female_Winner['Female_winner']==Proportion_Female_Winner['Female_winner'].max()]
max_female_decade_category
```

```python
max_female_dict={max_female_decade_category['decade'].values[0]:max_female_decade_category['category'].values[0]}
max_female_dict
```

```python
ax1=sns.catplot(x='category',y='Female_winner',data=Proportion_Female_Winner,
            kind='bar',col='decade',
            col_wrap=5, palette="PuRd")
ax1.fig.suptitle('The Nobel Prize Female', y=1.02)
```

```python
ax3=sns.relplot(x='decade',y='Female_winner',data=Proportion_Female_Winner,
                kind='line',hue='category',style='category',markers=True, 
                dashes=False)
ax3.fig.suptitle('Decade vs Proportion Female Winner', y=1.02)
```
